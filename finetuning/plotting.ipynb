```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
metrics_df = pd.read_csv("mistral_training_metrics.csv")
plt.figure(figsize=(8, 5))
plt.plot(metrics_df["epoch"], metrics_df["loss"], marker='o')
plt.title("Training Loss per Epoch")
plt.xlabel("Epoch")
plt.ylabel("Loss")
plt.grid(True)
plt.xticks(metrics_df["epoch"])
plt.tight_layout()
plt.savefig("mistral_training_loss.png")
plt.show()
```
